```python
#QUESTION 4: LARGEST PALINDORME MADE BY MULTIPLYING 2, 3 DIGIT NUMBERS
```

```python
def is_palindrome(number):
    # Convert the number to a string
    str_number = str(number)
    
    # Reverse the string
    reversed_str_number = str_number[::-1]
    
    # Check if the original string is equal to the reversed string
    return str_number == reversed_str_number

largestPalindrome = 0
for i in range(0, 1000):
    for j in range(0, 1000):
        if is_palindrome(i*j):
            print(i, j, i*j)
            if i*j > largestPalindrome:
                largestPalindrome = i*j

print(largestPalindrome)
```

```python
# QUESTION 10: sum of all prime numbers less than 2 million
```

```python
def sum_of_primes_below(limit):
    # Sieve of Eratosthenes
    sieve = [True] * limit
    sieve[0] = sieve[1] = False  # 0 and 1 are not prime numbers
    for start in range(2, int(limit**0.5) + 1):
        if sieve[start]:
            for multiple in range(start*start, limit, start):
                sieve[multiple] = False
    # Summing all prime numbers
    return sum(num for num, is_prime in enumerate(sieve) if is_prime)

# Setting the limit to 2 million
limit = 2000000
sum_primes = sum_of_primes_below(limit)
print(sum_primes)
```

```python
# QUESTION 12: first triangle number to have 500 divisors, chatted
```

```python
def generate_triangle_numbers(n):
    triangle_numbers = []
    for i in range(1, n + 1):
        triangle_number = i * (i + 1) // 2
        triangle_numbers.append(triangle_number)
    return triangle_numbers

# Generate the first 1000 triangle numbers
first_1000_triangle_numbers = generate_triangle_numbers(1000)

def count_divisors(n):
    count = 0
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            count += 2 if i != n // i else 1
    return count

def find_triangle_number_with_divisors(min_divisors):
    triangle_number = 0
    i = 1
    while True:
        triangle_number += i
        if count_divisors(triangle_number) > min_divisors:
            return triangle_number
        i += 1

# Find the first triangle number with over 500 divisors
min_divisors = 500
triangle_number = find_triangle_number_with_divisors(min_divisors)
print(f"The first triangle number with over {min_divisors} divisors is {triangle_number}")
```

```python
# QUESTION 14, which starting number gives the longest collatz chaindef collatz_length(n, memo):
    original_n = n
    length = 0
    while n != 1:
        if n in memo:
            length += memo[n]
            break
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        length += 1
    memo[original_n] = length + 1
    return length + 1

def generate_collatz_lengths(limit):
    memo = {}
    lengths = {}
    for i in range(1, limit + 1):
        lengths[i] = collatz_length(i, memo)
    return lengths

# Generate the Collatz sequence lengths for numbers from 1 to 1,000,000
limit = 1000000
collatz_lengths = generate_collatz_lengths(limit)

# Output the lengths (optional, can be commented out if not needed)
# for number, length in collatz_lengths.items():
#     print(f"Number: {number}, Length: {length}")

# Save the results to a file (optional)
# with open('collatz_lengths.txt', 'w') as f:
#     for number, length in collatz_lengths.items():
#         f.write(f"Number: {number}, Length: {length}\n")
```
